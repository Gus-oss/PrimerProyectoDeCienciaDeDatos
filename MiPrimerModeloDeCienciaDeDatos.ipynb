Supongamos que una empresa está analizando las ventas de un producto en función de los meses del año. Queremos ajustar una línea recta que prediga las ventas mensuales basándonos en datos históricos. Utilizaremos el cálculo simbólico con SymPy para encontrar los parámetros de esta línea que minimizan el error cuadrático medio (MSE).
Datos de Ejemplo

Imaginemos que tenemos los siguientes datos de ventas mensuales (en miles de unidades) durante los primeros cinco meses del año:

    Instalación y carga de bibliotecas necesarias.
    Definición de datos.
    Cálculo simbólico utilizando SymPy.
    Visualización de resultados con Matplotlib.
    Conclusiones.

Instalación y Carga de Bibliotecas

```python
!pip install sympy matplotlib numpy

import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
sp.init_printing()
```

Definición de Datos

Definimos los datos de ventas mensuales (en miles de unidades) durante los primeros cinco meses del año:

```python
# Datos de ejemplo: Meses del año y ventas en miles de unidades
meses = np.array([1, 2, 3, 4, 5])
ventas = np.array([2, 3, 5, 4, 6])
```

Cálculo Simbólico utilizando SymPy

Definimos las variables simbólicas y la función de error cuadrático medio (MSE), calculamos las derivadas parciales y resolvemos para encontrar los parámetros óptimos:

```python
# Definimos las variables simbólicas
m, b = sp.symbols('m b')

# Definimos la función simbólica para el error cuadrático medio (MSE)
def mse(m, b, x_data, y_data):
    n = len(x_data)
    error = sum((y_data[i] - (m * x_data[i] + b))**2 for i in range(n))
    return error / n

# Calculamos el MSE simbólicamente
mse_expr = mse(m, b, meses, ventas)

# Derivada parcial del MSE con respecto a m y b
dmse_dm = sp.diff(mse_expr, m)
dmse_db = sp.diff(mse_expr, b)

# Solucionamos el sistema de ecuaciones para encontrar m y b
solutions = sp.solve([dmse_dm, dmse_db], (m, b))

# Evaluamos las soluciones
m_opt = solutions[m]
b_opt = solutions[b]

# Calculamos los valores ajustados de ventas
ventas_fit = m_opt * meses + b_opt
print(m_opt)
print(b_opt)
print(ventas_fit)
```

Visualización de Resultados con Matplotlib

Ahora visualizamos los datos de ventas originales y la línea ajustada usando gráficos de Matplotlib con colores llamativos:

```python
# Configuración de los gráficos con colores llamativos
plt.figure(figsize=(10, 6))
plt.scatter(meses, ventas, color='blue', label='Datos de Ventas', s=100, edgecolor='black')
plt.plot(meses, ventas_fit, color='red', linewidth=2, label='Línea Ajustada')

# Configuración del gráfico
plt.xlabel('Meses', fontsize=14, color='darkblue')
plt.ylabel('Ventas (en miles)', fontsize=14, color='darkblue')
plt.title('Ajuste de una Línea Recta a las Ventas Mensuales', fontsize=16, color='darkred')
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

Conclusiones

Basado en el análisis y la visualización de los datos de ventas, podemos concluir lo siguiente:

    Tendencia General: La línea ajustada muestra una tendencia positiva, indicando un aumento general en las ventas durante los primeros cinco meses del año.

    Precisión del Modelo: Aunque la línea ajustada no captura todas las variaciones en las ventas, proporciona una buena aproximación de la tendencia general, lo que puede ser útil para hacer predicciones a corto plazo.

    Uso Futuro: La empresa puede utilizar este modelo para prever las ventas en los próximos meses, ajustando sus estrategias de marketing y producción en consecuencia.
